```python
from sklearn.preprocessing import QuantileTransformer

qt = QuantileTransformer(output_distribution="uniform")
```

```python
quantile_features = qt.fit_transform(orig_features)
```

```python
quantile_features
```

```python
# Convert quantile features to DataFrame with proper index
quantile_features_df = pd.DataFrame(quantile_features, columns=orig_features.columns, index=orig_features.index)
# Add Instance Group column to the quantile features
quantile_features_df['Instance Group'] = IGsources.values
quantile_features_df
```

```python
# Melt the dataframe to long format for plotting
quantile_features_long = quantile_features_df.melt(
    id_vars=['Instance Group'],
    var_name='Feature',
    value_name='Percentile Rank'
)
quantile_features_long
```

```python
# Calculate median for each feature by Instance Group
median_by_group = quantile_features_long.groupby(['Feature', 'Instance Group'])['Percentile Rank'].median().unstack()
median_by_group
# Calculate signed difference (IG1 - IG2)
median_difference = median_by_group['IG1'] - median_by_group['IG2']
median_difference
# Sort features by signed difference in decreasing order
sorted_features = median_difference.sort_values(ascending=False).index.tolist()
sorted_features
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Create the boxplot
fig = px.box(
    quantile_features_long,
    x='Feature',
    y='Percentile Rank',
    color='Instance Group',
    category_orders={'Instance Group': ['IG1', 'IG2'],
                     'Feature': sorted_features},
    color_discrete_map={'IG1': '#636EFA', 'IG2': '#EF553B'}  # Blue for IG1, Red for IG2
)

# Update layout to match the style in the image
fig.update_layout(
    template='plotly_white',
    font=dict(size=20),
    xaxis_title='Feature',
    yaxis_title='Percentile Rank',
    yaxis=dict(tickformat='.0%'),  # Format y-axis as percentages
    height=1000,
    width=1400,
    legend=dict(
        title='Instance Group',
        yanchor="top",
        y=0.99,
        xanchor="right",
        x=0.99
    )
)

# Rotate x-axis labels for better readability
fig.update_xaxes(tickangle=45)

fig.show()
```

```python
# Save the figure
fig.write_image(knn_fpath + 'mcc/overlapping/instance_group_boxplots.png', engine='kaleido', width=1400, height=1000)
```
