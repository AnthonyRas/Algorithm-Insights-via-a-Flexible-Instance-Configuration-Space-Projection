```python
from typing import Literal
import panel as pn
import numpy as np
from icsa import ICSA, set_seeds
from classification_icsa import load_metadata, MCC, load_knn_metadata, load_sgd_metadata, get_confusion, get_performance_data

pn.extension('plotly')
pn.config.theme = 'dark'
set_seeds(0)
```

```python
# KNN #
knn_fpath = 'analysis/classification/knn/'  # filepath for analysis files

def fit_overlapping(metadata, model_fpath_overlapping: str, optim_dir: Literal['min', 'max'], save=True, validation_split=0.1, epochs=150, verbose=0, **kwargs):
    analysis_overlapping = ICSA(optim_dir, mode='overlay', **kwargs)  # ICSA with overlapping spaces
    analysis_overlapping.fit_show(model_fpath_overlapping, metadata, validation_split=validation_split, epochs=epochs, save=save, verbose=verbose)
    return analysis_overlapping# fit

## MCC ##
knn_metadata_mcc = load_knn_metadata(knn_fpath, confusion_function=MCC)
```

```python
knn_metadata_mcc
```

```python
knn_analysis_mcc = fit_overlapping(knn_metadata_mcc, knn_fpath + 'mcc/overlapping/', 'max', save=True, performance_metric_name='MCC')
```

```python
# KNN #
knn_fpath = 'analysis/classification/knn/'  # filepath for analysis files

def fit_concatenated(metadata, model_fpath_cat: str, optim_dir: Literal['min', 'max'], save=True, validation_split=0.1, epochs=150, verbose=0, **kwargs):
    analysis_cat = ICSA(optim_dir, mode='concatenate', **kwargs)  # ICSA with concatenated spaces
    analysis_cat.fit_show(model_fpath_cat, metadata, validation_split=validation_split, epochs=epochs, save=save, verbose=verbose)  # fit the projection models
    return analysis_cat

## MCC ##
knn_analysis_mcc = fit_concatenated(knn_metadata_mcc, knn_fpath + 'mcc/concatenated/', 'max', save=True, performance_metric_name='MCC')
```
